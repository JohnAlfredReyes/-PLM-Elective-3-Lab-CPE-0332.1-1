```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/orange.png')
cv2_imshow(img)
```

```python
#convert to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#display new image
cv2_imshow(img_gray)
```

```python
#resize
resized_img = cv2.resize(img, (250, 250))

#display new image
cv2_imshow(resized_img)
```

```python
#flip an image
img_flip = cv2.flip(img_gray, 0)

#display new image
cv2_imshow(img_flip)
```

```python
import cv2
import matplotlib.pyplot as plt

# Load image from file (replace 'image.jpg' with your image file path)
image = cv2.imread('/content/orange.png')

# Convert BGR to RGB (OpenCV loads images as BGR by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert RGB to grayscale
gray_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

# Display original and grayscale images using matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original RGB Image')

plt.subplot(1, 2, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')

plt.show()
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load your grayscale image (replace this with your image loading code)
img_gray = cv2.imread('/content/orange.png', cv2.IMREAD_GRAYSCALE)

# Get image dimensions
height, width = img_gray.shape[:2]

# Define the rotation angle in degrees
angle = 75

# Calculate rotation matrix
rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), angle, 1)

# Perform affine transformation
img_rotated = cv2.warpAffine(img_gray, rotation_matrix, (width, height))

# Display the rotated image
cv2_imshow(img_rotated)
```
